```python
# --- Normalize and group result_df by scenario_id ---

# Define mappings for Technology and Quality to normalize names
tech_map = {
    '4Gn1': 'LTE',
    '4Gn': 'LTE',
    '4GSRS': 'LTE',
    '5Gn3': '5G',
    '5Gn78': '5G',
    '3G': '3G',
    'WIFI-2.4GHz': 'WiFi',
    'WIFI-5GHz': 'WiFi',
}
quality_map = {
    'Bonne': 'Good',
    'TRES-Bonne': 'VeryGood',
    'OPTIMAL': 'Optimal',
    'Eco': 'Eco',
    'Auto': 'Auto',
    'HIGH': 'High',
    'MAX': 'Max',
    '720p': '720p',
    '480p': '480p',
    '360p': '360p',
}

# Lowercase device/platform and apply mappings
result_df['Device'] = result_df['Device'].str.lower()
result_df['Platform'] = result_df['Platform'].str.lower()
result_df['Technology'] = result_df['Technology'].map(tech_map)
result_df['Quality'] = result_df['Quality'].map(quality_map)
# Log all combinations of Device, Platform, Technology, Quality BEFORE mapping
print("\n--- Original combinations (before mapping) ---")
print(result_df[['Device', 'Platform', 'Technology', 'Quality']].drop_duplicates().to_string(index=False))

# Default mobility to 'stat'
result_df['Mobility'] = 'stat'

# Build scenario_id: device_tech_platform_quality_mobility
result_df['scenario_id'] = (
    result_df['Device'] + "_" +
    result_df['Technology'] + "_" +
    result_df['Platform'] + "_" +
    result_df['Quality'] + "_" +
    result_df['Mobility']
)

# Convert power columns to numeric in case they're not already
result_df['Avg RF Power W'] = pd.to_numeric(result_df['Avg RF Power W'], errors='coerce')
result_df['Avg BAT Power W'] = pd.to_numeric(result_df['Avg BAT Power W'], errors='coerce')

# Group by scenario_id and compute mean values
video_streaming_scenario_summary_df = result_df.groupby('scenario_id')[['Avg RF Power W', 'Avg BAT Power W']].mean().reset_index()

# Optional: Rename columns for frontend clarity
video_streaming_scenario_summary_df.rename(columns={
    'Avg RF Power W': 'Avg_RF_Power_W',
    'Avg BAT Power W': 'Avg_BAT_Power_W'
}, inplace=True)

# Compute energy per minute (Wh) and convert to joules
video_streaming_scenario_summary_df['E_BAT_Jm'] = video_streaming_scenario_summary_df['Avg_BAT_Power_W'] * 60
video_streaming_scenario_summary_df['E_RF_Jm'] = video_streaming_scenario_summary_df['Avg_RF_Power_W'] * 60

# Add zero placeholders for compatibility
video_streaming_scenario_summary_df['E_BB_Jm'] = 0
video_streaming_scenario_summary_df['E_PA_Jm'] = 0

# Reorder columns
cols = ['scenario_id', 'E_BAT_Jm', 'E_RF_Jm', 'E_BB_Jm', 'E_PA_Jm']
video_streaming_scenario_summary_df = video_streaming_scenario_summary_df[cols]
# Save to CSV for web app simulator
video_streaming_scenario_summary_df.to_csv("./website/server/video_streaming_scenario_summary_df.csv", index=False)

# Preview result
print(video_streaming_scenario_summary_df.head())
```
